```python
import pandas as pd 
import numpy as np
from pathlib import Path

```

```python
col_nums_value=[]
for x in range(5,269)
    col_nums_value.append(x)
```

```python
# Read in ZHVI
file_path = Path("resources/zillow_home_value_index.csv")
home_value_index = pd.read_csv(file_path, parse_dates=col_nums_value, infer_datetime_format=True)
home_value_index.head(10)
```

```python
home_value_index.dtypes
```

```python
# Drop RegionType and StateName
home_value_index.drop(columns=['RegionType', 'StateName'], inplace=True)
home_value_index.head()
```

```python
df_2 = pd.melt(home_value_index, id_vars = ['RegionName', 'RegionID', 'SizeRank'])
               
df_2
```

```python
df_2.rename(columns={'variable' : 'Date'})
```

```python
df_2.to_csv("home_value_index_new_data.csv")
```
